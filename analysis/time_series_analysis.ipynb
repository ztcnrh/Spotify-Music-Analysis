```python
# Dependecies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import rcParams
import matplotlib.animation as animation
```

## Using MatplotLib to Chart a DataFrame

```python
# Loading the data
file = '../cleaned_data/cleaned_data_by_year.csv'
year_df = pd.read_csv(file)
year_df.head()
```

```python
# Parse out common ranges inot it's own DataFrame
line_data = year_df[['year','acousticness','danceability','energy','instrumentalness','liveness','speechiness','valence']]
line_data

# label='Loudness') # Out of sync
# label='Tempo') # Out of sync
# label='Popularity') # Out of sync
```

## Plotting time series chart (Feature scores vs. Years)

```python
# Time series plot

plt.figure(figsize=(16,8), dpi=200)

plt.plot(line_data['year'], line_data.iloc[:,1], label='Acousticness')
plt.plot(line_data['year'], line_data.iloc[:,2], label='Danceability')
plt.plot(line_data['year'], line_data.iloc[:,3], label='Energy')
plt.plot(line_data['year'], line_data.iloc[:,4], label='Instrumentalness')
plt.plot(line_data['year'], line_data.iloc[:,5], label='Liveness')
plt.plot(line_data['year'], line_data.iloc[:,6], label='Speechiness')
plt.plot(line_data['year'], line_data.iloc[:,7], label='Valence')

plt.grid(True, color='k', linestyle=':', alpha=0.25)
plt.title("Audio Features vs. Year")
plt.xlabel("Year")
plt.ylabel("Rating")
# plt.xlim(min(line_data['year']), max(line_data['year']))
plt.legend(bbox_to_anchor=(1, 1))

# Save figure
plt.savefig('../images/audio_features_over_years.png')

plt.show()
```
